```python
import requests
def err(a, b):
    try:
        return float(requests.get(f'http://ramcdougal.com/cgi-bin/error_function.py?a={a}&b={b}',headers={'User-Agent':'MyScript'}).text)
    except:
        print('bad input')
        return 

```

```python
h = 1e-4
gamma = 0.1

def GD(a, b):
#     m  = float('inf')
    loss = err(a, b)
    trial = 0
    while trial < 2:
        a -= gamma*(err(a+h, b)-err(a, b))/h
        b -= gamma*(err(a, b+h)-err(a, b))/h
#         print(loss, err(a,b))
        if loss > err(a, b):
            loss = err(a, b)
            trial = 0
        else:
            trial+=1
#         print(a,b,trial)
    return loss, a, b
            
```
